
Модель Кронига-Пени описывает поведение электронов в кристаллической решетке с периодической структурой потенциала. Она основана на решении уравнения Шредингера для электрона в таком потенциале:

\begin{equation}
-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2} + V(x)\psi = E\psi
\end{equation}

где $\psi$ - волновая функция электрона, $V(x)$ - потенциал в точке $x$, $E$ - энергия электрона.

Для одномерного кристалла с потенциалом, заданным выше, уравнение Шредингера может быть решено аналитически. Решение имеет вид:

\begin{equation}
\psi_n(x) =
 \begin{cases}
   A_1 e^{ik_ncx} + A_2 e^{-ik_ncx},&\text{$nc<x<nc+a$}\\
   B_1 e^{q_nx} + B_2 e^{-q_nx},&\text{$(nc+a)<x<(n+1)c$}
 \end{cases}
\end{equation}

где $k_n = \sqrt{\frac{2mE}{\hbar^2}}$ - волновой вектор электрона в области свободного пространства, $q_n = \sqrt{\frac{2m(U-E)}{\hbar^2}}$ - волновой вектор электрона в области потенциальной ямы, $A_1$, $A_2$, $B_1$, $B_2$ - коэффициенты, определяемые из граничных условий на границах областей.

Зона проводимости и зона запрещенных значений энергии определяются из условия существования решения уравнения Шредингера. Для этого требуется, чтобы коэффициент $B_2$ был равен нулю, т.е. чтобы решение было ограничено в области потенциальной ямы. Это условие приводит к следующему уравнению для энергии электрона:

\begin{equation}
\cos(k_nc a) = \frac{k_n}{q_n}\cos(q_n a)
\end{equation}

Решая это уравнение численно для разных значений энергии $E$, можно построить зонную структуру кристалла и анализировать изменение ширины запрещенных зон в зависимости от параметров потенциала.

Для случая свободного электрона ($U=0$) имеем $q_n = i\sqrt{\frac{2mE}{\hbar^2}}$, т.е. волновой вектор в области потенциальной ямы является мнимым. Уравнение (3) принимает вид $\cos(k_nc a) = \cosh(k_n a)$. Решение этого уравнения даёт ширину первой зоны Бриллюэна $2\pi/c$ и ширину запрещенной зоны:

\begin{equation}
E_g = \frac{\hbar^2\pi^2}{2ma^2}
\end{equation}

Для случая запертого электрона ($U=\infty$) имеем $q_n = \sqrt{\frac{2m(U-E)}{\hbar^2}}$, т.е. волновой вектор в области потенциальной ямы является действительным. Уравнение (3) принимает вид $\cos(k_nc a) = \cos(q_n a)$. Решение этого уравнения даёт ширину первой зоны Бриллюэна $2\pi/c$ и ширину запрещенной зоны:

\begin{equation}
E_g = U - \frac{\hbar^2\pi^2}{2ma^2}
\end{equation}




Для анализа изменения ширины запрещенной зоны в зависимости от параметров потенциала, мы можем рассмотреть два крайних случая: когда электрон совершенно свободен и когда электрон заперт внутри одной потенциальной ямы.

Когда электрон совершенно свободен, потенциал V(x) равен нулю во всем кристалле. В этом случае, ширина запрещенной зоны равна нулю, так как энергетические уровни электронов непрерывны.

Когда электрон заперт внутри одной потенциальной ямы, потенциал V(x) равен константе внутри ямы и бесконечности вне ямы. В этом случае, ширина запрещенной зоны зависит от ширины ямы. Чем шире яма, тем меньше ширина запрещенной зоны.

Для промежуточных случаев, когда U и a изменяются, ширина запрещенной зоны будет зависеть от этих параметров. Обычно, при увеличении высоты барьера U , ширина запрещенной зоны увеличивается, а при увеличении ширины ямы a , ширина запрещенной зоны уменьшается.

В целом, изменение ширины запрещенной зоны зависит от параметров потенциала V(x) и может быть рассчитано с использованием метода Кронига-Пени.
